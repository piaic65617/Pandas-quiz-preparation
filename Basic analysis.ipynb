# Value_counts()

```python
import pandas as pd
```

```python
aspl=pd.read_csv('../data/olympics.csv',skiprows=4)
aspl
```

```python
#value counts give the unique value o that series its a method
aspl.Edition.value_counts() # counts the value of num of medals from most to least(default)
```

```python
#aspl.Gender.value_counts()
aspl.Gender.value_counts(ascending=True, dropna=False)
```

#  sort_values()
## sort the value in series as axis is 0 we are sorting along a column

```python
ath=aspl.Athlete.sort_values()
ath
```

```python
#sorting using multiple series o data
aspl.sort_values(by =['Edition','Athlete'])
```

# Q1.In which events did Jesse Owens wins the Gold Medal?

```python
jo=aspl[aspl.Athlete == 'OWENS, Jesse']
jo
```

```python
jo.Event.value_counts()
```

# Q2.Which Country(city) has won most men's Gold Medals in a single badminton over the years?(sort the result alphabatically by players name)


```python
mbg=aspl[(aspl.Gender=='Men') & (aspl.Medal=='Gold') & (aspl.Sport=='Badminton') & (aspl.Event == 'singles')]
mbg.sort_values(by='Athlete')
```

# Q3.Which three countries have won the most medals in recent years (from 1984 to 2008)?

```python
recy=aspl[aspl.Edition >=1984]
recy.NOC.value_counts().head(3)
```

#  Q4.Display the male gold medal winners for the 100m Track & Field sprint event over the years. List the results starting with the most recent. Show the Olympic City, Edition, Athlete and the country they represent.

```python
gme=aspl[(aspl.Gender == 'Men') & (aspl.Medal == 'Gold') & (aspl.Event == '100m')]
gme.sort_values('Edition', ascending=False)[['City','Edition','Athlete','NOC']]
```
